```python
import networkx as nx
```

```python
CA=nx.MultiGraph()
```

```python
import matplotlib.pyplot as plt
import itertools
authors={
    1:'DJENONTIN A',
    2:'AKOGBETO M',
    3:'GREEN M',
    4:'BEACH R',
    5:'CORBEL V',
    6:'CHANDRE F'
}

auth=[k for k in authors]
edges = itertools.combinations(auth,2)

for k in authors:
    CA.add_node(k, name=authors[k])
title="Malaria vectors resistance to insecticides in Benin: current trends and mechanisms involved"
CA.add_edges_from(edges, doi='10.1186/s12936-015-0757-5',title=title,journal='Parasit Vectors.',year=2015)
edges
nx.draw(CA)
plt.show()
# for i in range(0,len(auth)):
#     CA.add_node(auth[i],auth[i])
# dir(CA)
```
